```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/IndianWeatherRepository.csv")

print("Original Shape:", df.shape)
print("\nOriginal Columns:")
print(df.columns.tolist())

# Columns to remove
columns_to_drop = [
    # Duplicate unit conversions
    'temperature_fahrenheit',
    'wind_mph',
    'pressure_in',
    'precip_in',
    'feels_like_fahrenheit',
    'visibility_miles',
    'gust_mph',

    # Location identifiers
    'country',
    'location_name',
    'region',
    'timezone',

    # Duplicate time representation
    'last_updated',

    # Potential leakage
    'condition_text',

    # Astronomical variables
    'sunrise',
    'sunset',
    'moonrise',
    'moonset',
    'moon_phase',
    'moon_illumination'
]

# Drop only those columns that actually exist
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

print("\nShape after removing unnecessary attributes:", df.shape)
print("\nRemaining Columns:")
print(df.columns.tolist())

# Save the cleaned dataset
df.to_csv("IndianWeatherRepository_CloudCover_Cleaned.csv", index=False)

print("\nCleaned dataset saved as:")
print("IndianWeatherRepository_CloudCover_Cleaned.csv")
```
